## **A. Membaca dan Menampilkan Citra**

Pada tahap ini dilakukan pembacaan dua citra menggunakan OpenCV. Selanjutnya dilakukan konversi warna dari format BGR ke RGB agar citra dapat ditampilkan dengan benar menggunakan Matplotlib.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Upload 2 gambar
uploaded = files.upload()

# Ambil nama file yang diupload
file_names = list(uploaded.keys())

print("File yang diupload:")
print(file_names)

# File pertama = kucing
# File kedua = anjing
img1 = cv.imread(file_names[0])
img2 = cv.imread(file_names[1])

# Cek apakah berhasil dibaca
if img1 is None or img2 is None:
    raise Exception("Gambar gagal dibaca. Upload ulang file JPG/PNG.")

print("=== IMAGE 1 (KUCING) ===")
print("Shape :", img1.shape)
print("Data Type :", img1.dtype)
print("Min Pixel :", img1.min())
print("Max Pixel :", img1.max())

print()

print("=== IMAGE 2 (ANJING) ===")
print("Shape :", img2.shape)
print("Data Type :", img2.dtype)
print("Min Pixel :", img2.min())
print("Max Pixel :", img2.max())
```

```python
img1_rgb = cv.cvtColor(img1, cv.COLOR_BGR2RGB)
img2_rgb = cv.cvtColor(img2, cv.COLOR_BGR2RGB)

plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.imshow(img1_rgb)
plt.title("Image1 - Kucing")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(img2_rgb)
plt.title("Image2 - Anjing")
plt.axis("off")

plt.show()
```

# ANALISIS BAGIAN A

Berdasarkan hasil pembacaan citra menggunakan OpenCV, diperoleh dua citra yang berbeda yaitu citra kucing sebagai image1 dan citra anjing sebagai image2. Informasi ukuran citra (shape), tipe data, serta nilai minimum dan maksimum pixel berhasil ditampilkan. Informasi tersebut penting karena digunakan untuk mengetahui karakteristik awal citra sebelum dilakukan manipulasi.

Setelah dilakukan konversi BGR ke RGB, warna citra terlihat lebih sesuai dengan warna aslinya. Hal ini terjadi karena OpenCV membaca citra menggunakan format BGR sedangkan Matplotlib menggunakan format RGB.

Jawaban Pertanyaan

1. Mengapa warna citra dapat terlihat tidak sesuai jika langsung ditampilkan setelah dibaca dengan OpenCV?

Warna citra dapat terlihat tidak sesuai karena OpenCV membaca citra dalam format BGR (Blue, Green, Red), sedangkan Matplotlib menampilkan citra dalam format RGB (Red, Green, Blue). Akibatnya warna merah dan biru dapat tertukar.

2. Apa perbedaan BGR dan RGB?

Perbedaan utama BGR dan RGB terletak pada urutan kanal warnanya. BGR menggunakan urutan Blue-Green-Red sedangkan RGB menggunakan urutan Red-Green-Blue.

3. Mengapa penting memahami format warna sebelum melakukan manipulasi citra?

Karena kesalahan memahami format warna dapat menyebabkan hasil visualisasi dan pengolahan citra menjadi tidak sesuai dengan kondisi sebenarnya.

# **B. Resize Citra**

Pada tahap ini dilakukan perubahan ukuran (resize) pada citra kucing dan citra anjing. Resize diperlukan agar kedua citra memiliki ukuran yang sama sehingga dapat digunakan pada proses manipulasi citra berikutnya seperti image blending.

```python
img1_resize = cv.resize(img1_rgb, (600,400))
img2_resize = cv.resize(img2_rgb, (600,400))

print("Ukuran Kucing Sebelum Resize :", img1_rgb.shape)
print("Ukuran Kucing Sesudah Resize :", img1_resize.shape)

print()

print("Ukuran Anjing Sebelum Resize :", img2_rgb.shape)
print("Ukuran Anjing Sesudah Resize :", img2_resize.shape)
```

```python
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.imshow(img1_rgb)
plt.title("Kucing Sebelum Resize")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(img1_resize)
plt.title("Kucing Setelah Resize")
plt.axis("off")

plt.subplot(2,2,3)
plt.imshow(img2_rgb)
plt.title("Anjing Sebelum Resize")
plt.axis("off")

plt.subplot(2,2,4)
plt.imshow(img2_resize)
plt.title("Anjing Setelah Resize")
plt.axis("off")

plt.show()
```

# ANALISIS BAGIAN B

Pada tahap resize, kedua citra berhasil diubah ukurannya menjadi 600 × 400 piksel. Proses ini dilakukan agar ukuran kedua citra sama sehingga dapat digunakan pada proses image blending. Jika ukuran citra berbeda, maka OpenCV tidak dapat menggabungkan kedua citra karena jumlah piksel pada masing-masing citra tidak sama.

Hasil resize menunjukkan bahwa bentuk objek utama pada citra masih dapat dikenali dengan baik. Citra kucing tetap terlihat jelas dengan latar belakang persawahan, sedangkan citra anjing tetap mempertahankan detail bulu dan latar belakang rerumputan.

Jawaban Pertanyaan

1. Mengapa dua citra harus memiliki ukuran yang sama sebelum dilakukan image blending?

Karena image blending menggabungkan nilai piksel pada posisi yang sama dari dua citra. Jika ukuran berbeda maka proses penggabungan tidak dapat dilakukan.

2. Apa risiko resize terhadap kualitas citra?

Resize dapat menyebabkan hilangnya detail gambar, munculnya blur, atau distorsi terutama jika ukuran diperbesar secara berlebihan.

3. Apakah resize dapat mengubah proporsi objek dalam citra?

Ya. Jika rasio panjang dan lebar tidak dipertahankan, objek dapat terlihat lebih lebar atau lebih tinggi dibandingkan bentuk aslinya.

# **C. Image Blending**

Pada tahap ini dilakukan penggabungan citra kucing dan citra anjing menggunakan metode image blending dengan beberapa kombinasi nilai alpha dan beta.

```python
blend1 = cv.addWeighted(img1_resize, 0.2, img2_resize, 0.8, 0)
blend2 = cv.addWeighted(img1_resize, 0.5, img2_resize, 0.5, 0)
blend3 = cv.addWeighted(img1_resize, 0.8, img2_resize, 0.2, 0)

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(blend1)
plt.title("Alpha=0.2 Beta=0.8")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(blend2)
plt.title("Alpha=0.5 Beta=0.5")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(blend3)
plt.title("Alpha=0.8 Beta=0.2")
plt.axis("off")

plt.show()
```

# ANALISIS BAGIAN C

Image blending dilakukan untuk menggabungkan dua citra menjadi satu citra baru. Pada praktikum ini digunakan citra kucing sebagai image1 dan citra anjing sebagai image2. Nilai alpha digunakan sebagai bobot citra kucing, sedangkan beta digunakan sebagai bobot citra anjing.

Pada alpha 0.2 dan beta 0.8, citra anjing terlihat lebih dominan karena memiliki bobot yang lebih besar. Pada alpha 0.5 dan beta 0.5, kedua citra tampak seimbang dan saling menyatu. Sedangkan pada alpha 0.8 dan beta 0.2, citra kucing menjadi lebih dominan.

Jawaban Pertanyaan

1. Alpha dan beta menentukan tingkat dominasi masing-masing citra.

2. Jika alpha lebih besar maka citra kucing lebih dominan.

3. Jika alpha dan beta terlalu besar maka citra dapat menjadi terlalu terang.

4. Image blending banyak digunakan pada watermark, editing foto, dan efek visual.

# **D. Image Negative**

Transformasi negatif dilakukan dengan membalik nilai intensitas pixel menggunakan persamaan:

s = 255 - r

```python
img_negative = 255 - img1_resize

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img1_resize)
plt.title("Kucing Asli")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(img_negative)
plt.title("Kucing Negative")
plt.axis("off")

plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(img1_resize.ravel(), bins=256, range=[0,256])
plt.title("Histogram Asli")

plt.subplot(1,2,2)
plt.hist(img_negative.ravel(), bins=256, range=[0,256])
plt.title("Histogram Negative")

plt.show()
```

# ANALISIS BAGIAN D

Transformasi negatif menghasilkan citra yang memiliki intensitas kebalikan dari citra asli. Area yang sebelumnya terang menjadi gelap, sedangkan area yang sebelumnya gelap menjadi terang.

Pada citra kucing, warna bulu yang semula terang berubah menjadi lebih gelap dan area bayangan menjadi lebih terang. Histogram hasil negatif terlihat berlawanan dengan histogram citra asli karena distribusi intensitas pixel dibalik.

Jawaban Pertanyaan

1. Pixel terang berubah menjadi gelap.

2. Pixel gelap berubah menjadi terang.

3. Histogram citra negatif menjadi kebalikan histogram asli.

4. Teknik ini sering digunakan pada citra medis dan analisis dokumen.

# **E. Transformasi Logaritmik**

Transformasi logaritmik digunakan untuk meningkatkan detail pada area gelap citra.

```python
c = 255 / np.log10(1 + np.max(img1_resize))

img_log = c * np.log10(img1_resize + 1)
img_log = np.array(img_log, dtype=np.uint8)

print("Nilai Minimum Sebelum :", img1_resize.min())
print("Nilai Maksimum Sebelum :", img1_resize.max())

print()

print("Nilai Minimum Sesudah :", img_log.min())
print("Nilai Maksimum Sesudah :", img_log.max())
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img1_resize)
plt.title("Kucing Asli")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(img_log)
plt.title("Log Transform")
plt.axis("off")

plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(img1_resize.ravel(), bins=256, range=[0,256])
plt.title("Histogram Asli")

plt.subplot(1,2,2)
plt.hist(img_log.ravel(), bins=256, range=[0,256])
plt.title("Histogram Log")

plt.show()
```

# ANALISIS BAGIAN E

Transformasi logaritmik memberikan peningkatan yang lebih besar pada pixel dengan intensitas rendah dibandingkan pixel dengan intensitas tinggi. Akibatnya area gelap menjadi lebih terang dan detail yang sebelumnya sulit terlihat menjadi lebih jelas.

Pada citra kucing, area bayangan serta bagian latar yang lebih gelap terlihat lebih terang setelah transformasi dilakukan. Perubahan pada area terang relatif kecil dibandingkan area gelap.

Jawaban Pertanyaan

1. Area gelap menjadi lebih terang.

2. Area terang mengalami perubahan yang lebih kecil.

3. Transformasi logaritmik cocok untuk memperjelas detail pada area gelap.

4. Jika digunakan berlebihan, citra dapat menjadi terlalu terang.

# **F. Transformasi Gamma**

Transformasi gamma digunakan untuk mengontrol tingkat kecerahan citra secara fleksibel.

```python
gammas = [0.5, 1, 2, 2.5]

plt.figure(figsize=(12,10))

for i, gamma in enumerate(gammas):

    img_norm = img1_resize / 255.0

    img_gamma = np.power(img_norm, gamma)

    img_gamma = np.uint8(img_gamma * 255)

    plt.subplot(2,2,i+1)
    plt.imshow(img_gamma)
    plt.title(f"Gamma = {gamma}")
    plt.axis("off")

plt.show()
```

# ANALISIS BAGIAN F

Transformasi gamma menghasilkan perubahan kecerahan yang berbeda tergantung nilai gamma yang digunakan.

Gamma 0.5 menghasilkan citra yang lebih terang dibandingkan citra asli.

Gamma 1 menghasilkan citra yang sama dengan citra asli.

Gamma 2 menghasilkan citra yang lebih gelap.

Gamma 2.5 menghasilkan citra yang semakin gelap.

Pada citra kucing terlihat bahwa semakin besar nilai gamma maka intensitas pixel semakin menurun sehingga citra menjadi lebih gelap.

Jawaban Pertanyaan

1. Gamma kurang dari 1 membuat citra lebih terang.

2. Gamma sama dengan 1 tidak mengubah citra.

3. Gamma lebih dari 1 membuat citra lebih gelap.

4. Gamma correction digunakan untuk memperbaiki pencahayaan citra.

5. Gamma lebih fleksibel dibandingkan transformasi logaritmik.
